# Impacted Schools

major_impacts does three things:

1.   chart_measures : calculations of the upper and lower percentile values per year.
2.   impacted_years : school rows that are above or below the values in the chart_measures
3.   impacted_schools: The final output that uses the impacted_years to find the data from the schools across all the years.




```python
def major_impacts(df, percentile):
    # Making sure the upper and lower is higher and lower, respectively
    df = df.reset_index(drop=True)
    upper, lower = (percentile, round(1 - percentile, 2)) if percentile >= (1 - percentile) else (round(1 - percentile, 2), percentile)

    # For each column per year, get the literal values of the upper / lower by using the percents
    ident_measures = df.groupby('End Year')[['Money Difference per school', 'Proficiency Score']].quantile([upper, lower], numeric_only=True).reset_index()
    chart_measures = pd.pivot_table(ident_measures, index=['End Year'], columns = ['level_1'] )
    chart_measures.columns = ['Lower money', 'Upper money', 'Lower score', 'Upper score']

    impacted_years = pd.DataFrame()

    # Iterate over years, and apply the unique filters for each year
    for year in chart_measures.index:
        year_df = df[df['End Year'] == year].reset_index(drop=True)

        # Select rows based on quantile values
        # The output is a T/F column
        lower_money, upper_money, lower_score, upper_score = chart_measures.loc[(year)]

        if pd.notna(lower_money) and pd.notna(upper_money):
          money_mask = ~year_df['Money Difference per school'].between(lower_money, upper_money, inclusive='both')

        if pd.notna(lower_score) and pd.notna(upper_score):
          score_mask = ~year_df['Proficiency Score'].between(lower_score, upper_score, inclusive='both')

        # Uses the T/F column to select the rows with a T.
        impacted_years = pd.concat([impacted_years, year_df[money_mask], year_df[score_mask]]).drop_duplicates()

    # Takes the school code, and asks for all data from those years.
    school_codes = impacted_years['School Code'].unique()
    impacted_schools = df[df['School Code'].isin(school_codes)]

    return impacted_schools.reset_index(drop=True)
```

```python
# Create selected_df based on conditions
# For some reason, when there is a null value it's always counted as high or low.
df_filtered = df[(df['District Code'] == 275) & (df['Level Code'] == 2)]

# The high and low high schools in JCPS.
selected_df = major_impacts(df_filtered, 0.15)
```

```python
def sca_export(df, file_name):
  # Select the needed columns
  df = df.loc[:, ['End Year', 'District', 'School', 'School Code', 'Level',
                  'Reported Spending per student', 'Money Difference per school', 'Proficiency Score', 'Rating']]
  # Rename columns
  df.columns = ['End_year', 'District', 'School', 'School_code', 'Level',
                'Reported_spending_per_student', 'Adjusted_spending_difference', 'Test_score', 'Test_rating']

  path = f'/content/drive/MyDrive/Colab Data/KY EPSB/All Years/{file_name}.json'
  df.to_json(path, orient="records")

  return print("Done")
```
